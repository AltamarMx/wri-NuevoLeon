```python
f = "../data/001_raw/ENCEVI_2018/aireacond.csv"
aire = pd.read_csv(f,index_col=0)
folios_con_ac = aire.index.unique()

# 2. Subconjunto de viv con AC
viv_con_ac = vivienda.loc[ vivienda.index.isin(folios_con_ac) ]

# 3. Subconjunto de viv sin AC
viv_sin_ac = vivienda.loc[ ~vivienda.index.isin(folios_con_ac) ]

# 4. Comprobar tamaños
print("Con AC:", viv_con_ac[viv_con_ac.entidad==19].factor_sem.sum())
print("Sin AC:", viv_sin_ac[viv_sin_ac.entidad==19].factor_sem.sum())
```

## Carga archivo de Hernandez con consumos de energ'ia y región climática

## Cambios de nombres de claves por valores y nuevos nombres

```python
# Supongamos que tu dataframe se llama vivienda y la columna con el código es 'super_cons'
mapeo_m2 = {
    1: 30,
    2: 55,
    3: 75,
    4: 100,
    5: 150,
    6: 200,
    7: np.nan,
    9: np.nan
}
mapeo_estrato = {
    1: 'Bajo',
    2: 'Medio bajo',
    3: 'Medio alto',
    4: 'Alto'
}
mapeo_climas = {
    "Extremely hot":'Cálido extremoso',
    "Temperate": 'Templado',
    "Tropical": 'Tropical'
}

# Crear la nueva columna "m2" aplicando el diccionario de mapeo
vivienda['m2'] = vivienda['super_cons'].map(mapeo_m2)
vivienda['estrato'] = vivienda['est_socio'].map(mapeo_estrato)
vivienda['climate_region_name'] = vivienda['climate_region_name'].map(mapeo_climas)
vivienda['kwh_mes'] = (vivienda['bill_kwh_estimation'] / vivienda['bill_days']) * 30
vivienda['kwh_m2'] = vivienda['bill_kwh_estimation']/vivienda['m2']
vivienda['kwh_persona'] = vivienda['bill_kwh_estimation']/vivienda['tot_integ']
vivienda['kwh_m2persona'] = vivienda['bill_kwh_estimation']/vivienda['tot_integ']/vivienda['m2']

del vivienda['bill_days']
del vivienda['est_socio']
del vivienda['bill_kwh_estimation']
# del vivienda['super_cons']
vivienda.rename(columns={'climate_region_name':'clima'},inplace=True)
vivienda.rename(columns={'factor_sem':'factor'},inplace=True)
vivienda.rename(columns={'tot_integ':'personas'},inplace=True)


# mapeo_m2 = {
#     1: '30_54',
#     2: '55_74',
#     3: '75_99',
#     4: '100_149',
#     5: '150_199',
#     6: '200_',
#     7: np.nan,
#     9: np.nan
# }
# vivienda['m2'] = vivienda['super_cons'].map(mapeo_m2)
del vivienda['super_cons']

vivienda
```

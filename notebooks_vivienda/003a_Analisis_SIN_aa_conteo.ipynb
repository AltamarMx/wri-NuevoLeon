```python
import pandas as pd
import numpy as np
```

## Folio, factor_sem, est_socio

```python
f = '../data/001_raw/ENCEVI_2018/vivienda.csv'
columnas = ['folio','factor_sem','est_socio','region']
vivienda = pd.read_csv(f,usecols=columnas)
vivienda.set_index('folio',inplace=True)
print(vivienda.info())
print(vivienda.factor_sem.sum())
```

## folio, super_cons

```python
f = '../data/001_raw/ENCEVI_2018/encevi.csv'
columnas = ['folio',
            # 'tipo_viv',
            # 'total_niv',
            'super_cons',
            # 'super_ter','tot_cuart','electri','local_com',
            # 'cons_med1','cons_med2','inicia1','mes_inicia1',
            # 'inica2','mes_inicia2',
           ]
encevi = pd.read_csv(f,usecols=columnas)
columnas = encevi.columns
columnas
encevi.set_index('folio', inplace=True)
vivienda = vivienda.join(encevi,how='inner')
print(vivienda.info())
print(vivienda.factor_sem.sum())
```

```python
# Supongamos que tu dataframe se llama vivienda y la columna con el código es 'super_cons'
mapeo_m2 = {
    1: 30,
    2: 55,
    3: 75,
    4: 100,
    5: 150,
    6: 200,
    7: np.nan,
    9: np.nan
}
mapeo_estrato = {
    1: 'Bajo',
    2: 'Medio bajo',
    3: 'Medio alto',
    4: 'Alto'
}
mapeo_climas = {
    1:'Cálido extremoso',
    2: 'Templado',
    3: 'Tropical'
}

# Crear la nueva columna "m2" aplicando el diccionario de mapeo
vivienda['m2'] = vivienda['super_cons'].map(mapeo_m2)
del vivienda['super_cons']
vivienda['estrato'] = vivienda['est_socio'].map(mapeo_estrato)
del vivienda['est_socio']
vivienda['clima'] = vivienda['region'].map(mapeo_climas)
del vivienda['region']
vivienda.rename(columns={'factor_sem':'factor'},inplace=True)

vivienda
```

```python
vivienda.factor.sum()
```

```python
f = '../data/001_raw/ENCEVI_2018/aireacond.csv'
aa = pd.read_csv(f)
aa.set_index('folio',inplace=True)
aa
```

## Separamos entre aire acondicionado y no aire acondicionado

```python
total_viv = vivienda.factor.sum()
total_viv
```

```python
vivienda
```

```python
# hogares con AC (intersección)
vivienda_aa = vivienda[vivienda.index.isin(aa.index)]

# hogares sin AC (diferencia)
vivienda_naa = vivienda[~vivienda.index.isin(aa.index)]
```

```python
print(vivienda_aa.factor.sum())
print(vivienda_naa.factor.sum())
```

```python
(vivienda_aa.groupby(['m2','clima','estrato']).sum()/total_viv*100).sum()
```

```python
(vivienda_naa.groupby(['m2','clima','estrato']).sum()/total_viv*100).sum()
```

```python
porcentajes_naa = (vivienda_naa.groupby(['m2','clima','estrato']).sum()/total_viv*100).reset_index()
porcentajes_naa.rename(columns={'factor':'porcentaje'},inplace=True)
porcentajes_naa['viviendas'] = porcentajes_naa.porcentaje*total_viv/100
porcentajes_naa.to_parquet("../data/002_intermediate/porcentajes_naa.parquet")
porcentajes_naa.viviendas.sum()
```

```python
porcentajes_aa = (vivienda_aa.groupby(['m2','clima','estrato']).sum()/total_viv*100).reset_index()
porcentajes_aa.rename(columns={'factor':'porcentaje'},inplace=True)
porcentajes_aa['viviendas'] = porcentajes_aa.porcentaje*total_viv/100
porcentajes_aa.to_parquet("../data/002_intermediate/porcentajes_aa.parquet")
porcentajes_aa.viviendas.sum()
```

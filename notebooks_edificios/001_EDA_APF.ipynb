```python
import pandas as pd
from climas import clima_dict
```

```python
f = "../data/001_raw/APF/consumos2022-2024APF.xlsx"
apf = pd.read_excel(f)
# del apf["Contraseña"]
del apf["Zona T"]
del apf["TR"]
del apf["Propio"]
del apf["Contraseña"]
suma = apf.iloc[:, 6:].sum(axis=1).rename("emensual")/36
apf = pd.concat([apf.iloc[:, :6],suma],axis=1)
apf["emensualm2"] = apf.emensual/ apf["Area (m2)"]
apf
```

```python

rename_dict = {
    'Uso genérico':      'generico',
    'Uso específico':    'especifico',
    'Municipio':         'municipio',
    'Entidad Federativa':'entidad',
    'Area (m2)':         'area',
    'AA':                'aa',
    'Consumo':           'consumo',
    'Contraseña':           'id',
}

apf = apf.rename(columns=rename_dict)
apf["ent_mun"] = apf.entidad + ","+ apf.municipio
apf
```

```python

# Supongamos que tu DataFrame se llama apf
unique_generico = apf['generico'].unique().tolist()
unique_entidad = apf['entidad'].unique().tolist()
unique_entidad = [entidad.capitalize() for entidad in unique_entidad]
unique_bioclima = apf['bioclima'].unique().tolist()

# Alinear las longitudes (rellena con cadenas vacías)
n = max(len(unique_generico), len(unique_entidad), len(unique_bioclima))
unique_generico += [''] * (n - len(unique_generico))
unique_entidad  += [''] * (n - len(unique_entidad))
unique_bioclima += [''] * (n - len(unique_bioclima))

# Crear DataFrame y volcar a Markdown
df_unicos = pd.DataFrame({
    'generico':   unique_generico,
    'entidad':    unique_entidad,
    'bioclima':   unique_bioclima
})
df_unicos.index = range(1, len(df_unicos) + 1)

print(df_unicos.to_markdown(index=True))
```
